1. Download the data.txt file

2. Open the file “data.txt” in read mode and load the data into a List

```python
listdata=[]
file = open('data.txt','r')
for y in file.readlines():
   # if y isdigit():
        listdata.append(float(y))
print(listdata)
        
```


3. Verify that each value in the List is a float ranging between -100 and 100;

```python
# Your Solution
if(all(type(y) is float for y in listdata) and (-100 < y<100 for y in listdata) ):
    print("Each value in the list is a float ranging between -100 and 100 ")
else:
    print('Not all the values are float and range between -100 and 100')
```


4. Round the number to the integer digit and convert each value of the List into an Integer;

```python
# Your Solution
for y in listdata:
    round(y)
print(listdata)
```


5. Sort the list in descending order;

```python
# Your Solution
listdata.sort(reverse=True)
print(listdata)
```


6. Save the newly sorted list of integers into a “.csv” file with the name “H1_< id >.csv”, replacing < id > by your student id.
    For example H1_20209999.csv

```python
import csv
new=[]
for y in listdata:
    new.append(str(y))
with open ('H1_m20190795','a') as file:
    w=csv.writer(file)
    w.writerow(new)
    
```

7. Print the values below -32

8. Create a dictionary with three keys. The first holds a list of the positive values from the list obtained in step 5, the second key the zeros and the last one, all the negatives.

```python
positive=[]
zero=[]
negative=[]
for y in listdata:
    if y < 0:
        negative.append(y)
    elif y == 0:
        zero.append(y)
    else:
        positive.append(y)
dictionary = {'key1': positive, 'key2':zero, 'key3':negative}        
print(dictionary)
```
